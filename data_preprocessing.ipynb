```python
common_df = pd.merge(temp, dust.drop(columns=['지점', '지점명']), on='일시')

common_df
```

```python
common_df['일시'] = pd.to_datetime(common_df['일시'])
```

```python
common_df['연도'] = common_df['일시'].dt.year
common_df['월'] = common_df['일시'].dt.month
common_df['일'] = common_df['일시'].dt.day
common_df['시각'] = common_df['일시'].dt.hour
```

```python
common_df.head(30)
```

```python
def get_season(month):
    if month in [3, 4, 5]:
        return '봄'
    elif month in [6, 7, 8]:
        return '여름'
    elif month in [9, 10, 11]:
        return '가을'
    else:
        return '겨울'

common_df['계절'] = common_df['월'].apply(get_season)
```
